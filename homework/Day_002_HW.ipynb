### 審計領域(杜絕弊案、財報分析)

我覺得機器學習適合用在，需要從大量資料中做出判斷，並且不受過去的經驗影響，做出客觀理性的決策。就我目前所認識的領域中，我覺得審計領域非常適合透過機器學習來做查核。因為審計人員必須接觸大量的會計資料，並且從中判讀出有疑慮的地方，因為運用人力，很容易因為各種原因，像是偷懶、縱容、串通等原因，而未能發現弊端。另外，有了電腦幫助判斷，也不需透過抽樣的方式來審查資料，可以徹底地檢視每一筆資料，將發生弊案的機率降到最低。

審計領域本身就是一個必須存在且有很多商機的領域，也難怪會有四大會計事務所統治著這個領域。相信機器學習可以改變審計的生態，幫廣大勞工擺脫被時間和客戶壓榨的生活。另外，因為機器學習能幫助審計員節省很多例行性的審查工作，也能將人腦多花費在客製化公司審計，讓審計員更能為不同產業間挖掘洞見，更能找出可能發生弊案的地方。
總結來說機器學習可以幫助審計領域:
1. 降低工作壓力(節省成本)
2. 節省例行性工作的時間(增進效率)
3. 專注於挖掘更多客製化的洞見(創造利潤)
4. 徹底盤查，避免抽樣審查的風險(預防危機)

除此之外，做審計的分析方式，某種程度也適合用來分析商業決策，除了審計以外也許還能做顧問諮詢。
